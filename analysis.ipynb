### Best players in the game

```python
# sort overall from highest to lowest
ovr = df[['short_name', 'overall']].sort_values(by='overall', ascending=False).head(10)
ovr
```

### Wonder Kids

```python
# Wonder kids
wonder_kids = df[['short_name', 'potential', 'age']].sort_values(by='potential', ascending=False)
```

```python
#sort by age
wonder_kids = wonder_kids[wonder_kids['age'] < 25]
wonder_kids.head(10)
```

### Best players per position

```python
def best_position(position):
    df = players[players['team_position'] == position][['short_name', 'overall', 'team_position']]
    return df.sort_values(by='overall', ascending=False).head(15)
```

1. Best Strikers in the game

```python
best_position('ST')
```
