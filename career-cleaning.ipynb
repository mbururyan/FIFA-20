## Validity

#### Remove useless columns
#### Columns to be dropped :

- player url
- long name
- dob
- player_position
- real_face
- player_tags
- national jersey
- player traits
- some stats

```python
def delete_column(dataframe, col_name):
    dataframe.drop([col_name], axis=1, inplace=True)
    return dataframe.head()
```

```python
# drop player url
delete_column(df, 'player_url')
```

```python
# drop long name
delete_column(df, 'long_name')
```

```python
# drop dob
delete_column(df, 'dob')
```

```python
# drop player position
delete_column(df, 'player_positions')
```

```python
#drop remaining columns
delete_column(df, 'real_face')
delete_column(df, 'player_tags')
delete_column(df, 'nation_position')
delete_column(df, 'nation_jersey_number')
delete_column(df, 'player_traits')
```

```python
# delete last few columns
df.drop(['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'rcb', 'lcb', 'cb', 'rb'], axis=1, inplace=True)
df.head()
```

```python
delete_column(df, 'cm')
```

```python
df.drop(['loaned_from', 'joined', 'contract_valid_until'], axis=1, inplace=True)
```
